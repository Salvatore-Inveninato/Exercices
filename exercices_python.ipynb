## Exercice 1
Définissez une classe Satellite() qui permette d’instancier des objets simulant des satellites artificiels lancés dans l’espace, autour de la terre. Le constructeur de cette classe initialisera les attributs d’instance suivants, avec les valeurs par défaut indiquées : masse = 100, vitesse = 0.

Lorsque l’on instanciera un nouvel objet Satellite(), on pourra choisir son nom, sa masse et sa vitesse.
Les méthodes suivantes seront définies :
- impulsion(force, duree) permettra de faire varier la vitesse du satellite. Pour savoir comment, rappelez-vous votre cours de physique : la variation de vitesse Δvsubie par un objet de masse m soumis à l’action d’une force F pendant un temps t vaut Δv= (F * t)/m. Par exemple : un satellite de 300 kg qui subit une force de 600 Newtons pendant 10 secondes voit sa vitesse augmenter (ou diminuer) de 20 m/s.
- affiche_vitesse() affichera le nom du satellite et sa vitesse courante.
- energie() renverra au programme appelant la valeur de l’énergie cinétique du satellite. Rappel : l’énergie cinétique se calcule à l’aide de la formule Ec= (m * v²)/2

```python
class Satellite:
    def __init__(self, nom, masse=100, vitesse=0):
        self.nom = nom
        self.masse = masse
        self.vitesse = vitesse

    def  impulsion(self, force, duree):
        self.vitesse += (force * duree)/self.masse

    def affiche_vitesse(self):
        print(f"Le satellite {self.nom} a une vitesse de {self.vitesse} m/s.")

    def energie (self):
        return (self.masse * self.vitesse**2)/2

satellite1 = Satellite("Satellite 1", 200, 1000)
satellite1.affiche_vitesse()
satellite1.impulsion(600, 10)
satellite1.affiche_vitesse()
print(f"L'energie cinetique du satellite 1 est de {satellite1.energie()} Joules.")
```

## Exercice 2

Écrivez une fonction distance() qui permet de calculer la distance entre deux points. (Il faudra vous rappeler le théorème de Pythagore !). Cette fonction attendra évidemment deux objets Point() comme arguments.

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

class Distance:
    def __init__(self, point_1, point_2):
        self.point_1 = point_1
        self.point_2 = point_2
    
    def distance(self):
        return ((self.point_2.x - self.point_1.x)**2 + (self.point_2.y - self.point_1.y)**2)**0.5

point_a = Point(4,1)
point_b = Point(12,150)
distance_a = Distance(point_1=point_a,point_2=point_b)
distance_a.distance()
```

## Exercice 4

Créez une classe "Etudiant" :

    - Un étudiant aura également une variable contenant la moyenne de ses notes ; 
    - Ecrivez les méthodes get / set pour tous ces attributs (NB : la liste de notes devra pouvoir être accessible en totalité ou pour un seul élément (fourni en paramètre));

    - Ecrivez une méthode AVG qui calcule la moyenne des notes de l’étudiant (NB : tout ajout / modification / suppression dans la liste contenant les notes des étudiants doit mettre à jour la valeur de la moyenne) ;

```python
class Etudiant:
    def __init__(self, noma, nom, prenom, age):
        self.noma = noma
        self.nom = nom
        self.prenom = prenom

        if age > 100 or age < 12:
            self.age = 18
        else : 
            self.age = age

        self.liste_notes = []
    
    def __str__(self):
        return f"L'élève est {self.prenom} {self.nom}, a {self.age} et porte le numéro de noma {self.noma}"

    def affiche_noma(self):
        print(f"Le noma de {self.prenom} {self.nom} est {self.noma}.")

    def affiche_nom_prenom(self):
        print(f"L'étudiant s'appelle {self.prenom} {self.nom}.")

    def affiche_age(self):
        print(f"L'étudiant à {self.age} ans.")

    def affiche_liste_notes(self):
        if len(self.liste_notes) == 0:
            print("L'étudiant n'a pas encore de notes, la moyenne ne peut donc pas être calculée.")
        
        else: 
            print(self.liste_notes)
    
    def affiche_avg(self):
        if len(self.liste_notes) == 0:
            print("L'étudiant n'a pas encore de notes, la moyenne ne peut donc pas être calculée.")
        
        else:
            num = 0
            denom = 0
            for i,j in self.liste_notes:
                num += i
                denom += j
            
            print(f"La moyenne actuelle de {self.prenom} {self.nom} est de {round(num/denom * 20, 2)}.")

    def definir_noma(self, noma):
        self.noma = noma

    def definir_age(self, age):
        self.age = age

    def definir_nom_prenom(self,nom = None, prenom=None):
        if nom is not None:
            self.nom = nom
        if prenom is not None:
            self.prenom = prenom

    def ajouter_note(self, num, denom):
        self.liste_notes.append((num,denom))

etu1 = Etudiant(noma= 123456789, nom= "Dupont", prenom = "Martin", age = 25)

etu1.affiche_age()
print(etu1)
etu1.affiche_noma()
etu1.affiche_nom_prenom()
etu1.affiche_age()
etu1.affiche_liste_notes()
etu1.affiche_avg()

etu1.ajouter_note(20,30)
etu1.affiche_liste_notes()
etu1.affiche_avg()

etu1.definir_age(45)
etu1.affiche_age()

etu1.definir_nom_prenom(prenom="John")
etu1.affiche_nom_prenom()

etu1.definir_noma(noma=987654321)
etu1.affiche_noma()

etu1.definir_nom_prenom(nom="Dujardin", prenom="Bill")
etu1.affiche_nom_prenom()

etu1.ajouter_note(90,100)
etu1.affiche_liste_notes()
etu1.affiche_avg()
```

## Exercice 6

Créez une classe Appareil.  
- Un appareil est caractérisé par une marque (visibilité protected). 
- Un appareil possède la fonction allumer, qui ne renvoie rien et ne prend pas de paramètre (corps de la fonction vide).  
- La classe ne pourra pas être instanciée. 

Créez une classe Telephone, héritant de la classe Appareil.  
- Son constructeur initialise l’attribut marque passé en paramètre.   
- La fonction allumer affichera: « dringggg ». 

Créez une classe Ordinateur, héritant de la classe Appareil.  
- Son constructeur initialise l’attribut marque passé en paramètre. 
- Un ordinateur possède également une caractéristique « taille » de type « int ». 
- La fonction allumer affichera: « Vive les PCs ». 
Tester ensuite le tout en créant une liste contenant 2 objets de chaque type d’appareil (c’est-à
dire des classes Telephone et Ordinateur), et en appelant la fonction allumer sur chacun de ces 
4 appareils. 
Respecter tous les principes de l’orienté objet.

```python
from abc import ABC, abstractmethod

class Appareil(ABC):
    def __init__(self,marque):
        self.marque = marque

    @abstractmethod
    def allumer(self):
        pass

class Telephone(Appareil):
    pass

    def allumer(self):
        print("dringggg")

class Ordinateur(Appareil):
    def __init__(self, marque : str, taille : int):
        self.taille = taille
        super().__init__(marque)

    def allumer(self):
        print("Vive les PCs")
        
tel1 = Telephone("Apple")
ordi1 = Ordinateur("Apple",79)
tel2 = Telephone("Huawei")
ordi2 = Ordinateur("HP", 66)

liste_appareils = [tel1, ordi1, tel2, ordi2]

for i in liste_appareils:
    i.allumer()
```
